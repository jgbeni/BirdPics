# Download BirdPics dataset in Google Drive

This notebook follows the code tutorial provided in https://www.kaggle.com/discussions/general/74235. It is **required** that you have already downloaded your Kaggle's **API key** (*kaggle.json*) to your computer.

We start by importing the relevant packages and mounting Google Drive:

```python
import os
from google.colab import drive,files
import zipfile
drive.mount('/content/drive')
```

Now we install the Kaggle package:

```python
! pip install -q kaggle
```

Now browse and upload the kaggle.json file from your computer

```python
files.upload()  # Upload the kaggle.json file
```

We now move the key to a kaggle folder and change the permissions

```python
! mkdir ~/.kaggle
! cp kaggle.json ~/.kaggle/
! chmod 600 ~/.kaggle/kaggle.json
```

We are finally ready to download the **BirdPics dataset**:

```python
!kaggle datasets download -d jgbeniqu/birdpics-spanish-migratory-bird-image-dataset
```

We create a BirdPics folder in your personal Google Drive to store the dataset

```python
drive_dir = '/content/drive/MyDrive/BirdPics'  # Update this path to your specific folder in Google Drive
os.makedirs(drive_dir, exist_ok=True)
os.makedirs(drive_dir+'/data',exist_ok=True)
```

And finally we unzip the dataset into the BirdPics folder...

```python
zip_ref = zipfile.ZipFile('birdpics-spanish-migratory-bird-image-dataset.zip', 'r')
zip_ref.extractall(drive_dir+'/data')
zip_ref.close()
```
